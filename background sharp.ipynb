```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('bird background.jpg',0)# read the gray scale image 
hist = cv2.calcHist([img],[0],None,[256],[0,256])#we are calculating the histogram for this img 
ret,l = cv2.threshold(img,75,255,cv2.THRESH_BINARY)#threshold for gray scale img and the output in plot 1
plt.hist(img.ravel(),256,[0,256]); plt.show()#drowing the histogram

img_inv=cv2.bitwise_not(l)#we used not operation to make the background =1 and the object =0 

img_bb=cv2.bitwise_and(img_inv,img)#we used and operation to show image has background without object 

img_ab = cv2.blur(img_bb,(9,9))#blurring on the img_bb .the output will be image has background blurred without object



img_obj=cv2.bitwise_and(l,img)#and operation between the image which has object=1 , background =0 >>>with grayscale img 
#output will be image has object without background 


img_final=cv2.add(img_ab,img_obj)#add the image which has blurred background with image which has an object 
#output will be image has an object with blurred background




#drowing the image when we are making threshold on it 
plt.title('threshold')
plt.xticks([]),plt.yticks([])
plt.imshow(l,'gray')
plt.show()




#drawing the image when we are making not operation 
plt.imshow(img_inv,'gray')
plt.title('inv')
plt.xticks([]),plt.yticks([])
plt.show()


#drawing the image which has only background
plt.imshow(img_bb,'gray')
plt.title('bb')
plt.xticks([]),plt.yticks([])
plt.show()



#drawing the image when we are making blurring on the backround 
plt.imshow(img_ab,'gray')
plt.title('ab')
plt.xticks([]),plt.yticks([])
plt.show()


#drawing the image which has only object
plt.imshow(img_obj,'gray')
plt.title('onj')
plt.xticks([]),plt.yticks([])
plt.show()

#drawing the final output 
plt.imshow(img_final,'gray')
plt.title('ab')
plt.xticks([]),plt.yticks([])
plt.show()






#show the orginal image
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
