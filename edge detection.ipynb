```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('flower2.jpg',0)
l = cv2.Laplacian(img,0)
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5) #kernal size is 5*5
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)

img3=cv2.add(img,l)

img4=l-img


plt.subplot(1,4,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,2),plt.imshow(l,cmap = 'gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,3),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])
plt.subplot(1,4,4),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])
plt.show()

plt.imshow(img3,cmap = 'gray')
plt.title('add'), plt.xticks([]), plt.yticks([])
plt.show()



plt.imshow(img4,cmap = 'gray')
plt.title('sub'), plt.xticks([]), plt.yticks([])
plt.show()
```
