```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img =cv2.imread('duck.pgm',0)
ret,imgf1 =cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
th1 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
ad1 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

img2 =cv2.imread('flower.pgm',0)
ret1,imgf2 =cv2.threshold(img2, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
th2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
ad2 = cv2.adaptiveThreshold(img2,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

img3 =cv2.imread('tools.pgm',0)
ret2,imgf3 =cv2.threshold(img3, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
th3 = cv2.adaptiveThreshold(img3,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)
ad3 = cv2.adaptiveThreshold(img3,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

plt.subplot(1,5,1), plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,2), plt.hist(img.ravel(), 256)
plt.axvline(x=ret, color='r',linestyle='dashed',  linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,3), plt.imshow(imgf1,cmap = 'gray')
plt.title('Otsu'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,4), plt.imshow(th1, 'gray')
plt.title('Ad Mean'), plt.xticks([]), plt.yticks([])
plt.subplot(1,5,5), plt.imshow(ad1, 'gray')
plt.title('AD Gaussian'), plt.xticks([]), plt.yticks([])
plt.show()

plt.subplot(2,5,1), plt.imshow(img2,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(2,5,2), plt.hist(img2.ravel(), 256)
plt.axvline(x=ret1, color='r',linestyle='dashed',  linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
plt.subplot(2,5,3), plt.imshow(imgf2,cmap = 'gray')
plt.title('Otsu'), plt.xticks([]), plt.yticks([])
plt.subplot(2,5,4), plt.imshow(th2, 'gray')
plt.title('Ad Mean'), plt.xticks([]), plt.yticks([])
plt.subplot(2,5,5), plt.imshow(ad2, 'gray')
plt.title('Ad Gaussian'), plt.xticks([]), plt.yticks([])
plt.show()


plt.subplot(3,5,1), plt.imshow(img3,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(3,5,2), plt.hist(img3.ravel(), 256)
plt.axvline(x=ret2, color='r',linestyle='dashed',  linewidth=2)
plt.title('Histogram'), plt.xticks([]), plt.yticks([])
plt.subplot(3,5,3), plt.imshow(imgf3,cmap = 'gray')
plt.title('Otsu'), plt.xticks([]), plt.yticks([])
plt.subplot(3,5,4), plt.imshow(th3, 'gray')
plt.title('Ad Mean'), plt.xticks([]), plt.yticks([])
plt.subplot(3,5,5), plt.imshow(ad3, 'gray')
plt.title('Ad Gaussian'), plt.xticks([]), plt.yticks([])
plt.show()
```
